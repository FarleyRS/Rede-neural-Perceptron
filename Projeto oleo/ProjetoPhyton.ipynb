```python
import pandas as pd
import numpy as np

#Ler arquivo
#df = pd.read_excel('/content/tabela.xlsx')
df = pd.read_excel('/content/tabelateste.xlsx')
```

```python
#entradas
#Y =np.array(df['d'])

x1 = np.array(df['x1'])
x2 = np.array(df['x2'])
x3 = np.array(df['x3'])

X = np.vstack((x1,x2,x3))

# Bias
bias = 1
```

```python
#Define o numero de epocas e o numero de amostras (q)
numEpocas = 0
#q = len(Y)
q=10
#Taxa de aprendizado
eta = 0.01

#Define o vetor de pesos aleatorios
W = np.random.random([1, 4])
Wr = W
#Array para armazenar os erros
#e = np.zeros(len(Y))
e = np.zeros(10)
```

```python
numEpocas = 0
def funcaoAtivacao(valor):
  # A função de ativação a degrau bipolar
  if valor< 0.0:
    return(-1)
  else:
    return(1)

r=True
while r :
  numEpocas+=1
  for k in range(q):
    #Insere o bias no vetor de entrada
    Xb = np.hstack((bias,X[:,k]))

    #calcular o campo induzido
    V = np.dot(W, Xb)

    #calcular a saida o perceptron
    Yr = funcaoAtivacao(V)

    #calcular o erro: e = ( Y - Yr)
    e[k] = Y[k]  - Yr

    #treinamento do perceptron
    W = W + eta*e[k]*Xb
    if len(set(e))!=1 and numEpocas<=2000:
      r=True
    else:
      r=False
    print("Resposta linha ",k,":",Yr)
  print("fim da ",numEpocas," epoca")
print("Vetor de erros (e) = "+str(e))
```
